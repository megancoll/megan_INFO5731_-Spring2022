```python
# You code here (Please add comments in the code):

# These are my imports. I include BeautifulSoup to parse thru the data and pandas to create a nice table.

import requests
from bs4 import BeautifulSoup 
import pandas as pd

import csv
from collections import defaultdict
```

```python

# making the agent
user_agent = ({'User-Agent':
			'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
			AppleWebKit/537.36 (KHTML, like Gecko) \
			Chrome/90.0.4430.212 Safari/537.36',
			'Accept-Language': 'en-US, en;q=0.5'})

# setting up to parse
def get_page_contents(url):
    page = requests.get(url, headers = user_agent)
    return BeautifulSoup(page.text, 'html.parser')

# setting a dict
hotels_dictionary = defaultdict(dict)

# iterating thru each page
for offset in range(0, 1000, 30):
    print('--- page offset:', offset, '---')

    # url link to TripAdvisor
    url = 'https://www.tripadvisor.com/Hotels-g295424-oa' + str(offset) + '-Dubai_Emirate_of_Dubai-Hotels.html#EATERY_LIST_CONTENTS' 

    # making soup variable
    soup = get_page_contents(url) 

    # finding hotel names, review counts, and ratings
    for i, link in enumerate(soup.find_all('a', {'property_title'})):
        num = offset + i
        output = ''.join(c for c in link.text if not c.isnumeric() and c != ".")
        hotels_dictionary[num]["Hotel Name"] = output.lstrip()

    for i, review in enumerate(soup.findAll('a',{'class':'review_count'})):
      num = offset + i
      hotels_dictionary[num]["Review Count"] = review.text.strip()

    for i, rating in enumerate(soup.findAll('a',{'class':'ui_bubble_rating'})):
      num = offset + i
      hotels_dictionary[num]["Rating"] = rating['alt']
```

```python
# returning the pandas table
df1 = pd.DataFrame.from_dict(hotels_dictionary)
tdf1 = df1.T

# getting 1000 samples of my data
tdf1.head(1000)
```

```python
def acm_scrape(base_url):
    """
    Code adapted from example function sigir_proceedings. This function works with a search page on the acm database.
    Since it is not conference proceedings, there are no sections to deal with, just search results. Returns dataFrame
    for one page of search results. For results from multiple pages, function must be called on each page. 
    """
    # driver
    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    # create Dataframe for data with columns
    comments = pd.DataFrame(
        columns=['article_type', 'article_title', 'article venue', 'article_author', 'article_year', 'article_id',
                  'article_abstract', 'total_citation', 'total_download', 'pdf_link'])
    driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
    driver.get(base_url)
    articles = driver.find_elements(By.CLASS_NAME, 'issue-item-container')
    for article in driver.find_elements(By.CLASS_NAME, 'issue-item-container'):
      # article type information
      article_type = article.find_element(By.CLASS_NAME, 'issue-heading').text
      # print(article_type)
      if article_type == 'SECTION':
          continue
      elif article_type == "RESEARCH-ARTICLE" or article_type == "ARTICLE":
          # article meta-information
          article_information = article.find_element(By.CLASS_NAME, 'issue-item__content').text.split('\n')
          # print(article_information)
          # article title
          article_title = article_information[0]
          # article author
          article_author = article_information[1]
          # article venue --> can be found in article information
          article_venue = article_information[2]
          # article year
          if len(re.findall(r'2\d\d\d', article_information[2])) > 0:
              article_year = re.findall(r'2\d\d\d', article_information[2])[0]
              article_year = int(article_year)
          else:
              article_year = ''
          # print(article_year)

          # article doi
          if len(re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+',
                            article_information[2])) > 0:
              article_url = \
                  re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+',
                              article_information[2])[0]
          else:
              article_url = ''
          # print(article_url)
          # abstract
          if len(article_information)>3:
              article_abstract = article_information[3]
          else:
              article_abstract = ''
          # citation count
          if len(article_information)>4:
              total_citation = article_information[4]
          else:
              total_citation = ''
          # download count
          if len(article_information)>5:
              total_download = article_information[5]
          else:
              total_download = ''
          # download link
          try:
              pdf_link = article.find_element(By.CLASS_NAME, 'btn--icon.simple-tooltip__block--b.red.btn')
              if pdf_link != None:
                  download_link = pdf_link.get_attribute('href')
                  # print(download_link)
              else:
                  download_link = ''
          except NoSuchElementException:
              download_link = ''
          comments.loc[len(comments)] = [article_type, article_title, article_venue, article_author,article_year,article_url,article_abstract,total_citation,total_download,download_link]
    # put comments Dataframe into results csv
    comments.to_csv('search_sigir_article_information.csv')
    # create and return DataFrame from results csv (still keep the csv just in case)
    res_df = pd.read_csv("search_sigir_article_information.csv")
    return res_df

```

```python
# loop through first 60 pages to get 1000 research articles
for i in range(80):
  megan_url = f"https://dl.acm.org/action/doSearch?fillQuickSearch=false&target=advanced&expand=dl&AfterYear=2012&BeforeYear=2022&AllField=information+retrieval&startPage={i}&pageSize=50"
  if i == 0:
    full_df = acm_scrape(megan_url)
  else:
    old_df = full_df
    new_df = acm_scrape(megan_url)
    frames = [old_df, new_df]
    full_df = pd.concat(frames)

# drop dummy column
final_df = full_df.drop(["Unnamed: 0"], axis=1)
```
