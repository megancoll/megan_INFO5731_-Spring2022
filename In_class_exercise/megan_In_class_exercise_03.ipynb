```python
#Import

import pandas as pd
```

```python
#Reading in file

twitter_data = pd.read_csv('Tweets.csv')

twitter_data
```

```python
#Cleaning the dataset of a row that has null values

tw_data = twitter_data.dropna()

tw_data
```

```python
#Location specific of data I want to pull from the pandas

txt = tw_data['text']

txt
```

```python
#Imports needed

import re

import nltk
nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')

from nltk import word_tokenize
```

```python
#This is the function for the adjectives feature

def adj_count(text):
  j_count = 0

  tokens = nltk.word_tokenize(text)
  tagged = nltk.pos_tag(tokens)

  new_t = [word + "_" + tag for (word, tag) in tagged]
 
#Calculating the total number of POS for the adjectives

  tag_list = [word.split("_")[1] for word in new_t]

  for tag in tag_list:
    if tag:
      if tag[0] == "J":
        j_count += 1

  return j_count
```

```python
#This is the function for the negation feature

def neg_count(text):
  neg_count = 0
  t = word_tokenize(text)
  for word in t:
    if word.lower() in ["no", "not", "n't", "nt"]:
      neg_count += 1
  return neg_count
```

```python
#Intensity words list

ins_list = [
  "absolutely",
  "amazingly",
  "astoundingly",
  "awful",
  "bitterly",
  "bloody",
  "completely",
  "crazy",
  "dead",
  "dreadfully",
  "colossally",
  "especially",
  "exceptionally",
  "excessively",
  "extremely",
  "extraordinarily",
  "fantastically",
  "frightfully",
  "fully",
  "hella",
  "incredibly",
  "insanely",
  "literally",
  "mad",
  "mightily",
  "outrageously",
  "particularly",
  "phenomenally",
  "precious",
  "quite",
  "radically",
  "rather",
  "real",
  "really",
  "remarkably",
  "ridiculously",
  "so",
  "somewhat",
  "strikingly",
  "super",
  "supremely",
  "terribly",
  "terrifically",
  "too",
  "totally",
  "unbelievably",
  "veritable",
  "very"
]
```

```python
#This is the function for the intensity feature

def ins_count(text):
  ins_count = 0
  t = word_tokenize(text)
  for word in t:
    if word.lower() in ins_list:
      ins_count += 1
  return ins_count
```

```python
#This is the function for the elongation feature

def el_count(text):
  t = word_tokenize(text)

  regex = re.compile(r"(.)\1{2}") #This regex code catches letters that have repeated more than twice
  el_count = len([word for word in t if regex.search(word)])
  
  return el_count
```

```python
#This is the function for the special characters feature

def spc_count(text):
  sp_char = """!@#$%^&*()-+?_=,<>/"""
  
  spc_count = 0
  for c in text:
    if c.lower() in sp_char:
      spc_count += 1
  return spc_count
```

```python
#Applying all the listed features in a for loop

all_features = []
for r in txt.index:
  this_text_features = []
  this_text = txt.loc[r]

  num_adj = adj_count(this_text)
  num_negation = neg_count(this_text)
  num_ins = ins_count(this_text)
  num_el = el_count(this_text)
  num_spc = spc_count(this_text)
  this_text_features = [num_adj, num_negation, num_ins, num_el, num_spc]
  all_features.append(this_text_features)

all_features
```

Question 3 (10 points): Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)." Select the most important features you extracted above, rank the features based on their importance in the descending order. 

I chose the document frequency method (DF). The summary from the paper is copied below.

*Document frequency method (DF) : document frequency of a term is defined as the total
number of documents in the document collection that contain the term. The basic idea of
DF is that rare terms are considered non-informative for classification and they should be
removed during feature selection [51, 88].*

```python
#Needed import

from collections import defaultdict
```

```python
#The Document Frequency checks which features are most common across the documents

#Intuition: Features that only occur rarely, are less informative than very common features.

informative_counts = defaultdict(int)

for r in txt.index:
  this_text = txt.loc[r]
  num_adj = adj_count(this_text)
  if num_adj > 0:
    informative_counts["Num Adjs"] += 1 
  num_negation = neg_count(this_text)
  if num_negation > 0:
    informative_counts["Num Negation"] += 1
  num_ins = ins_count(this_text)
  if num_ins > 0:
    informative_counts["Num Intensity"] += 1
  num_el = el_count(this_text)
  if num_el > 0:
    informative_counts["Num Elongated"] += 1
  num_spc = spc_count(this_text)
  if num_spc > 0:
    informative_counts["Num Special Char"] += 1

ordered_features = sorted(informative_counts, key=informative_counts.get, reverse=True)

for i in range(len(ordered_features)):
  num = str(i + 1)
  print(num + ": " + ordered_features[i] + ", Count of " + str(informative_counts[ordered_features[i]]))
```
